# Hypothesis Testing

Hypothesis Test 1: Deposit Type and Cancellation (from Q5)

Business question: Does requiring a Non Refundable deposit reduce cancellation rates
compared to No Deposit?

Null hypothesis (H0): There is no association between deposit type and cancellation
status. The cancellation rate is the same across all deposit types.

Alternative hypothesis (H1): There is a significant association between deposit type and
cancellation status. Specifically, Non Refundable deposits have lower cancellation rates
than No Deposit bookings.

Why I chose this test: Both variables are categorical. Deposit type has three categories
(No Deposit, Non Refundable, Refundable). Cancellation status has two categories
(canceled, not canceled). The chi-square test of independence determines whether the
two categorical variables are related.

Statistical test: Chi-square test of independence
α = 0.05

```python
import scipy.stats as st
cont = pd.crosstab(house["deposit_type"],house["is_canceled"])
temp = st.chi2_contingency(cont)
print(temp)
chi2_stat, p_value, dof, expected = temp
print("Chi-squared statistic:", chi2_stat)
print("p-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. There is a significant association between Deposit Type and Cancellation.")
else:
    print("Result: Fail to reject the null hypothesis. There is no significant association between Deposit Type and Cancellation")
```

# Hypothesis 2 : Do Online Traveling Agency bookings have significantly different lead times compared to Direct bookings?

Hypothesis	Statement

H₀ (Null)	There is no difference in mean lead time between OTA and Direct bookings

H₁ (Alternative)	There is a significant difference in mean lead time between OTA and Direct bookings

We Use here 2- sample T test


```python
Online = house[house['market_segment'] == 'Online TA']['lead_time']
Direct = house[house['market_segment'] == 'Direct']['lead_time']


# Perform independent two-sample t-test
t_stat, p_value = st.ttest_ind(Online, Direct, equal_var=False) 
print("t statistic:", t_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print('Result: Reject the null hypothesis. There is a statistically significant difference in average Lead time between customers on online TA and Direct.')
    print('Online TA lead time mean:', Online.mean(), ',\n Direct lead time mean:', Direct.mean())
else:
    print('Result: Fail to reject the null hypothesis. There is no statistically significant difference in average Lead time between customers on online TA and Direct.')
```
